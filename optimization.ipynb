```python
import fluxback as fb
import numpy as np
import matplotlib.pyplot as plt
```

```python
from fluxback.optimizer import SGD
```

```python
f = lambda a, b: a*a + b*b

x = fb.Tensor(5.0, requires_grad=True)
y = fb.Tensor(4.0, requires_grad=True)

optim = SGD([x, y], lr=0.1)

xs = []
ys = []
for _ in range(10):
	xs.append(x.values[()])
	ys.append(y.values[()])
	z = f(x, y)
	z.zero()
	z.back()
	optim.step()
```

```python
X = np.linspace(-10, 10, 100)
Y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(X, Y)

Z = X*X + Y*Y

plt.contour(X, Y, Z, np.logspace(-1, 2, 20))
plt.axis('equal')

plt.plot(xs, ys, 'o-', label='SGD')
plt.legend()
```

# Rosenbrock

```python
def rosenbrock(x, y):
	return (1-x)**2 + 6*(y-x**2)**2
```

```python
xx = np.linspace(-4, 4, 1000)
yy = np.linspace(-1, 3, 1000)
XX, YY = np.meshgrid(xx, yy)
ZZ = rosenbrock(XX, YY)

plt.contour(XX, YY, ZZ, np.logspace(-0.5, 2, 10))
plt.show()
```
